```python
total = 53
dropouts = 2
dismissed = 21

if total < dropouts:
    print("Too many dropouts!")

if total > dropouts:
    print("Not enough dropouts! We need to dismiss some!")

if total < dismissed:
    print("The course is finished!")

if total > dismissed:
    print("The course rolls on!")

dismissed +=20

if total >= dismissed:
    print("Total is greater than or equal to dismissed.")

if total <= dismissed:
    print("Total is less than or equal to dismissed.")

if total == dismissed:
    print("Everyone is dismissed from the course.")
```
