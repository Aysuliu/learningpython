```python
import csv 
x = []
y = []

with open('sales.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        x.append(int(row[0]))
        y.append(int(row[1]))
        
plt.plot(x,y)
plt.xlabel('Year')
plt.ylabel('number of dvd sales')
plt.title('annual number of sales from 2002 to 2011')
plt.grid()
plt.plot(x,y, 'b--')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt 
x = 0, 10, 15, 25, 35, 50
y1 = [5*i for i in x] 
y2 = [2*i**2 for i in x]

plt.title(" y1 = x, y2 = x^2") 
plt.xlabel("x")         
plt.ylabel("y1, y2")    
plt.grid()              
plt.plot(x, y1, 'y--')
plt.plot(x, y2, 'g.-')  
```

```python
import matplotlib.pyplot as plt 
x = 1, 3, 5, 7, 9
y1 =[6*i for i in x]
y2 = [i**2 for i in x]

plt.figure(figsize=(9, 12))
plt.subplot(2, 1, 1)
plt.plot(x, y1, 'b-')               
plt.title(": y1 = x, y2 = x^2") 
plt.ylabel("y1", fontsize=15) 
plt.grid(True)                
plt.subplot(2, 1, 2)
plt.plot(x, y2,'p--')               
plt.xlabel("x", fontsize=12)  
plt.ylabel("y2", fontsize=15) 
plt.grid(True)                
```

```python
import matplotlib.pyplot as plt
x=[1,3,5,7,9]
y1=[1,2,4,6,8]
y2=[3,6,9,12,15]
plt.figure(figsize=(12,7))
plt.plot(x,y1, 'o-r', alpha=0.7, label='cats', lw=2, mec='g', mew=5, ms=7)
plt.plot(x,y2, 'v-.g', label='dogs', mec='m', lw=2, mew=2, ms=10)
plt.grid()
plt.legend()
plt.show()
```
